### ML/DL Algorithms — The Catalog

> You don't need to memorize these. But you should know the categories and when each applies.

#### Supervised Learning — "Learn from labeled examples"

> You give the model input-output pairs. It learns the mapping. Then predicts on new inputs.

| Algorithm | Type | What It Does | ML or DL? | When to Use | Project |
|-----------|------|-------------|-----------|-------------|---------|
| **Linear Regression** | Regression | Fits a straight line to predict continuous values | ML | Predicting prices, scores, quantities | P1 |
| **Logistic Regression** | Classification | Predicts probability of a class (despite the name, it's classification) | ML | Binary yes/no decisions, baseline model | P1 |
| **Decision Tree** | Both | Splits data into branches based on feature values | ML | Explainable decisions, small datasets | P1 |
| **Random Forest** | Both | Ensemble of many decision trees (votes on answer) | ML | Tabular data, robust baseline — **start here** | P1 |
| **XGBoost / LightGBM** | Both | Gradient-boosted trees — #1 on Kaggle for tabular data | ML | Structured/tabular data, competitions | P1 |
| **SVM** (Support Vector Machine) | Both | Finds the best boundary between classes | ML | Small-to-medium datasets, text classification | P1 |
| **KNN** (K-Nearest Neighbors) | Both | Classifies by finding K closest examples | ML | Simple baseline, recommendation | P1 |
| **Naive Bayes** | Classification | Probability-based, assumes feature independence | ML | Text classification, spam filters | P1 |
| **Neural Network (MLP)** | Both | Layers of connected neurons learn complex patterns | DL | When classical ML plateaus, large datasets | P2 |
| **CNN** (Convolutional NN) | Classification | Specialized for grid-like data (images, spatial) | DL | Image classification, object detection | P2 |
| **RNN / LSTM** | Sequence | Processes sequential data (text, time series) | DL | Time series, older NLP (replaced by transformers) | P2 |
| **Transformer** | Sequence | Attention-based architecture — powers GPT, BERT, Claude | DL | NLP, LLMs, machine translation, everything modern | P2, P4 |

#### Unsupervised Learning — "Find hidden patterns in unlabeled data"

> No labels. The model discovers structure on its own.

| Algorithm | What It Does | ML or DL? | When to Use | Example |
|-----------|-------------|-----------|-------------|---------|
| **K-Means** | Groups data into K clusters by proximity | ML | Customer segmentation, document grouping | Group users by behavior |
| **DBSCAN** | Density-based clustering (finds irregular shapes) | ML | Anomaly detection, spatial data, noise-tolerant | Find fraud clusters |
| **Hierarchical Clustering** | Builds a tree of nested clusters | ML | Taxonomy building, gene analysis | Organize products |
| **PCA** (Principal Component Analysis) | Reduces dimensions while preserving variance | ML | Visualization, feature reduction, speed up training | Compress 100 features → 10 |
| **t-SNE / UMAP** | Non-linear dimension reduction for visualization | ML | Visualizing high-dimensional data in 2D/3D | Plot word embeddings |
| **Autoencoder** | Neural net that compresses then reconstructs data | DL | Anomaly detection, feature learning, denoising | Detect unusual transactions |
| **GAN** (Generative Adversarial Network) | Two networks compete: one generates, one discriminates | DL | Image generation, data augmentation | Generate synthetic images |

#### Reinforcement Learning — "Learn by trial and error"

> An agent takes actions in an environment, gets rewards/penalties, learns optimal behavior.

| Algorithm | What It Does | When to Use | Example |
|-----------|-------------|-------------|---------|
| **Q-Learning** | Learns value of actions in states (table-based) | Simple environments, discrete actions | Game AI, routing |
| **DQN** (Deep Q-Network) | Q-Learning with neural networks | Complex environments, Atari games | Game-playing AI |
| **PPO** (Proximal Policy Optimization) | Stable policy gradient method | Robotics, complex continuous control | ChatGPT's RLHF training |
| **RLHF** (RL from Human Feedback) | Trains LLMs to align with human preferences | LLM fine-tuning | How GPT/Claude learn to be helpful |

### ML vs DL — When to Use Which

| Factor | Classical ML | Deep Learning |
|--------|-------------|---------------|
| **Data size** | Works with 100s to 10,000s of rows | Needs 10,000s to millions of rows |
| **Feature engineering** | **YOU** design features (domain expertise required) | Network **learns** features automatically |
| **Training time** | Seconds to minutes | Minutes to days (GPU recommended) |
| **Hardware** | CPU is fine | GPU/TPU for anything serious |
| **Explainability** | Often interpretable (decision trees, coefficients) | Black box (hard to explain why) |
| **Structured/tabular data** | **ML wins** — Random Forest, XGBoost dominate | DL often worse on tabular data |
| **Images, text, audio** | ML struggles (needs manual feature extraction) | **DL wins** — CNNs, transformers dominate |
| **When to start** | **Always start here** — fast, interpretable, good baseline | When ML plateaus or data is unstructured |

```
                        Structured Data (CSV, tables)     Unstructured Data (images, text, audio)
                        ────────────────────────────      ──────────────────────────────────────
Best approach:          Random Forest / XGBoost (ML)      CNN / Transformer (DL)
Start with:             Logistic Regression (baseline)    Pre-trained model (transfer learning)
Our project:            P1                                P2, P3, P4
```

> **Rule of thumb:** Start with classical ML. If it's good enough, ship it. If not, try DL.  
> **Exception:** Text/image/audio → go straight to DL (pre-trained transformers/CNNs).

### Same Problem, ML vs DL Approach

| Problem | ML Approach | DL Approach |
|---------|------------|-------------|
| **Spam detection** | Extract features (word counts, sender, length) → Random Forest | Feed raw text → Transformer (BERT) auto-learns features |
| **Image classification** | Extract features (edges, colors, histograms) → SVM | Feed raw pixels → CNN auto-learns edges, shapes, objects |
| **Customer churn** | Engineer features (tenure, usage, complaints) → XGBoost | Same features → Neural Network (DL rarely better here) |
| **Language translation** | Feature engineering + statistical models (old way) | Seq-to-seq Transformer (modern way — GPT, Claude) |
| **Fraud detection** | Transaction features → Random Forest | Transaction sequences → LSTM or Transformer |

> **Key insight:** DL's superpower is **automatic feature learning** — it figures out what matters.  
> ML's superpower is **working well with small data** and being **explainable**.  
> In production, many companies use **both** — ML for tabular, DL for text/images.

````python
# ============================================================
# HELLO WORLD: Supervised ML — Classification (7 algorithms)
# ============================================================
# All 7 algorithms on the SAME dataset so you can compare them head-to-head.
# Dataset: Iris (150 flowers, 4 features, 3 species)

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

# Load and split data
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(
    iris.data, iris.target, test_size=0.3, random_state=42
)
print(f"Dataset: {len(iris.data)} samples, {iris.data.shape[1]} features, {len(iris.target_names)} classes")
print(f"Classes: {list(iris.target_names)}")
print(f"Train: {len(X_train)}, Test: {len(X_test)}\n")

# Run ALL 7 classifiers — same data, same split
classifiers = {
    "Logistic Regression": LogisticRegression(max_iter=200),
    "Decision Tree":       DecisionTreeClassifier(random_state=42),
    "Random Forest":       RandomForestClassifier(n_estimators=100, random_state=42),
    "SVM":                 SVC(kernel='rbf'),
    "KNN (K=5)":           KNeighborsClassifier(n_neighbors=5),
    "Naive Bayes":         GaussianNB(),
}

print(f"{'Algorithm':<22} {'Accuracy':>8}  {'Correct':>7}  Notes")
print(f"{'-'*65}")

results = {}
for name, clf in classifiers.items():
    clf.fit(X_train, y_train)               # train
    predictions = clf.predict(X_test)        # predict
    acc = accuracy_score(y_test, predictions) # evaluate
    correct = sum(predictions == y_test)
    results[name] = acc
    
    note = ""
    if name == "Random Forest":
        note = "← usually best for tabular data"
    elif name == "Logistic Regression":
        note = "← always start here (baseline)"
    
    print(f"  {name:<20} {acc:>7.1%}  ({correct}/{len(y_test)})  {note}")

best = max(results, key=results.get)
print(f"\nBest on this dataset: {best} ({results[best]:.1%})")

# Show what a single prediction looks like
sample = X_test[0].reshape(1, -1)
rf = classifiers["Random Forest"]
pred = rf.predict(sample)[0]
print(f"\nSingle prediction example:")
print(f"  Features: {sample[0]}  (sepal_len, sepal_wid, petal_len, petal_wid)")
print(f"  Predicted: {iris.target_names[pred]}")
````

````python
# ============================================================
# HELLO WORLD: Supervised ML — Regression (predict a number)
# ============================================================
# Regression = predict a continuous value (price, salary, temperature)
# vs Classification = predict a category (spam/not spam, species)

import numpy as np
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score

# Generate simple data: house size → price
np.random.seed(42)
sizes = np.random.uniform(800, 3000, 100)  # sq ft
prices = sizes * 150 + np.random.normal(0, 20000, 100)  # ~$150/sqft + noise

X = sizes.reshape(-1, 1)  # sklearn expects 2D input
y = prices

# Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Run 3 regression algorithms
regressors = {
    "Linear Regression":   LinearRegression(),
    "Ridge Regression":    Ridge(alpha=1.0),
    "Random Forest":       RandomForestRegressor(n_estimators=100, random_state=42),
}

print("Predicting house prices from square footage")
print(f"{'Algorithm':<22} {'MAE':>10} {'R² Score':>10}  Notes")
print(f"{'-'*60}")

for name, reg in regressors.items():
    reg.fit(X_train, y_train)
    predictions = reg.predict(X_test)
    mae = mean_absolute_error(y_test, predictions)
    r2 = r2_score(y_test, predictions)
    
    note = ""
    if name == "Linear Regression":
        note = "← baseline"
    elif name == "Random Forest":
        note = "← captures non-linearity"
    
    print(f"  {name:<20} ${mae:>9,.0f}  {r2:>9.3f}  {note}")

# Show a prediction
lr = regressors["Linear Regression"]
sample_size = np.array([[2000]])  # 2000 sq ft
predicted_price = lr.predict(sample_size)[0]
print(f"\nLinear Regression prediction:")
print(f"  House size: {sample_size[0][0]:.0f} sq ft → Predicted price: ${predicted_price:,.0f}")
print(f"  Coefficient: ${lr.coef_[0]:,.0f}/sq ft  (learned rate)")
print(f"  Intercept:   ${lr.intercept_:,.0f}  (base price)")
print(f"  Formula:     price = {lr.coef_[0]:.0f} × size + {lr.intercept_:.0f}")
````

````python
# ============================================================
# HELLO WORLD: Unsupervised ML — K-Means, PCA, DBSCAN
# ============================================================
# No labels! The algorithm discovers structure on its own.

import numpy as np
from sklearn.cluster import KMeans, DBSCAN
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

iris = load_iris()
X = iris.data  # 4 features, no labels used for training

# ---- K-Means Clustering: group data into K clusters ----
print("=== K-Means Clustering ===")
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X)

# Compare discovered clusters to actual species
print(f"  Found {len(set(cluster_labels))} clusters")
print(f"  Cluster centers shape: {kmeans.cluster_centers_.shape}")
print(f"\n  Cluster assignments vs actual species:")
for cluster_id in range(3):
    mask = cluster_labels == cluster_id
    actual = iris.target[mask]
    # Which species dominates this cluster?
    from collections import Counter
    species_counts = Counter(actual)
    dominant = iris.target_names[species_counts.most_common(1)[0][0]]
    print(f"    Cluster {cluster_id}: {sum(mask)} samples → mostly {dominant} {dict(species_counts)}")

print(f"\n  K-Means discovered the species WITHOUT seeing any labels!")

# ---- PCA: reduce 4 dimensions → 2 for visualization ----
print("\n=== PCA (Dimensionality Reduction) ===")
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X)

print(f"  Original shape:   {X.shape}  (150 samples × 4 features)")
print(f"  Reduced shape:    {X_2d.shape}  (150 samples × 2 components)")
print(f"  Variance kept:    {sum(pca.explained_variance_ratio_):.1%}")
print(f"    Component 1:    {pca.explained_variance_ratio_[0]:.1%}")
print(f"    Component 2:    {pca.explained_variance_ratio_[1]:.1%}")
print(f"  → 2 dimensions capture {sum(pca.explained_variance_ratio_):.1%} of the information!")

# ---- DBSCAN: density-based clustering (finds outliers) ----
print("\n=== DBSCAN (Density-Based Clustering) ===")
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

dbscan = DBSCAN(eps=0.8, min_samples=5)
db_labels = dbscan.fit_predict(X_scaled)

n_clusters = len(set(db_labels)) - (1 if -1 in db_labels else 0)
n_noise = list(db_labels).count(-1)
print(f"  Found {n_clusters} clusters + {n_noise} noise points (outliers)")
print(f"  → DBSCAN doesn't need K specified. It finds clusters by density.")
print(f"  → Points labeled -1 are outliers (don't belong to any cluster)")
print(f"  Cluster sizes: {dict(Counter(db_labels))}")

print("\n--- Key Differences ---")
print("  K-Means: YOU specify K (number of clusters). Finds round clusters.")
print("  DBSCAN:  Algorithm decides K. Finds irregular shapes. Detects outliers.")
print("  PCA:     Not clustering — reduces dimensions for visualization or speed.")
````

````python
# ============================================================
# HELLO WORLD: Deep Learning — Neural Network (PyTorch)
# ============================================================
# Same Iris classification — but now with a neural network.
# Compare: 3 lines of sklearn (above) vs ~30 lines of PyTorch (below).
# The tradeoff: more code, but can learn FAR more complex patterns.

try:
    import torch
    import torch.nn as nn
    import torch.optim as optim
    from sklearn.datasets import load_iris
    from sklearn.model_selection import train_test_split
    from sklearn.preprocessing import StandardScaler
    import numpy as np

    # Prepare data (same Iris dataset)
    iris = load_iris()
    X_train, X_test, y_train, y_test = train_test_split(
        iris.data, iris.target, test_size=0.3, random_state=42
    )
    scaler = StandardScaler()
    X_train_s = scaler.fit_transform(X_train)
    X_test_s = scaler.transform(X_test)

    # Convert to PyTorch tensors
    X_train_t = torch.FloatTensor(X_train_s)
    y_train_t = torch.LongTensor(y_train)
    X_test_t = torch.FloatTensor(X_test_s)
    y_test_t = torch.LongTensor(y_test)

    # ---- Define the neural network ----
    class IrisNet(nn.Module):
        """Simple feedforward neural network: 4 → 16 → 8 → 3"""
        def __init__(self):
            super().__init__()
            self.layers = nn.Sequential(
                nn.Linear(4, 16),     # 4 input features → 16 neurons
                nn.ReLU(),            # activation function
                nn.Linear(16, 8),     # 16 → 8 neurons
                nn.ReLU(),
                nn.Linear(8, 3)       # 8 → 3 output classes
            )
        
        def forward(self, x):
            return self.layers(x)

    model = IrisNet()
    criterion = nn.CrossEntropyLoss()        # loss function for classification
    optimizer = optim.Adam(model.parameters(), lr=0.01)  # optimizer

    total_params = sum(p.numel() for p in model.parameters())
    print(f"Neural Network: 4 → 16 → 8 → 3  ({total_params} parameters)")
    print(f"Training for 100 epochs...\n")

    # ---- Training loop ----
    for epoch in range(100):
        # Forward pass
        outputs = model(X_train_t)
        loss = criterion(outputs, y_train_t)
        
        # Backward pass + optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        if (epoch + 1) % 25 == 0:
            with torch.no_grad():
                test_outputs = model(X_test_t)
                _, predicted = torch.max(test_outputs, 1)
                acc = (predicted == y_test_t).sum().item() / len(y_test_t)
                print(f"  Epoch {epoch+1:3d}: loss={loss.item():.4f}, test_accuracy={acc:.1%}")

    # Final evaluation
    model.eval()
    with torch.no_grad():
        test_outputs = model(X_test_t)
        _, predicted = torch.max(test_outputs, 1)
        acc = (predicted == y_test_t).sum().item() / len(y_test_t)

    print(f"\n  Final accuracy: {acc:.1%}")
    print(f"\n--- sklearn (3 lines) vs PyTorch (~30 lines) ---")
    print(f"  sklearn:  model.fit(X, y) → model.predict(X_test)  [done]")
    print(f"  PyTorch:  define network → loss → optimizer → training loop → evaluation")
    print(f"  PyTorch gives you CONTROL (custom architectures, GPU, gradients)")
    print(f"  sklearn gives you SPEED (fast experiments, less code)")

except ImportError:
    print("PyTorch not installed. Run: pip install torch")
    print("Or open this notebook in Google Colab (pre-installed).")
````

````python
# ============================================================
# HELLO WORLD: Deep Learning — CNN (Image Classification Concept)
# ============================================================
# CNNs are specialized neural networks for images/spatial data.
# This shows the architecture — full image training comes in P2.

try:
    import torch
    import torch.nn as nn

    # ---- CNN Architecture: how it works ----
    class SimpleCNN(nn.Module):
        """CNN that could classify 28×28 grayscale images (like handwritten digits)."""
        def __init__(self):
            super().__init__()
            # Convolutional layers — learn to detect edges, shapes, patterns
            self.conv_layers = nn.Sequential(
                nn.Conv2d(1, 16, kernel_size=3, padding=1),  # 1 channel in → 16 filters
                nn.ReLU(),
                nn.MaxPool2d(2),                              # 28×28 → 14×14
                nn.Conv2d(16, 32, kernel_size=3, padding=1), # 16 → 32 filters
                nn.ReLU(),
                nn.MaxPool2d(2),                              # 14×14 → 7×7
            )
            # Fully connected layers — classify based on learned features
            self.fc_layers = nn.Sequential(
                nn.Flatten(),                  # 32 × 7 × 7 → 1568
                nn.Linear(32 * 7 * 7, 64),
                nn.ReLU(),
                nn.Linear(64, 10),             # 10 digit classes (0-9)
            )
        
        def forward(self, x):
            x = self.conv_layers(x)
            x = self.fc_layers(x)
            return x

    model = SimpleCNN()
    total_params = sum(p.numel() for p in model.parameters())
    
    # Simulate a batch of 4 grayscale images (28×28)
    fake_images = torch.randn(4, 1, 28, 28)  # batch=4, channels=1, 28×28
    output = model(fake_images)

    print("=== CNN (Convolutional Neural Network) ===")
    print(f"  Architecture: Conv(1→16) → Pool → Conv(16→32) → Pool → FC(1568→64→10)")
    print(f"  Total parameters: {total_params:,}")
    print(f"\n  Input shape:  {fake_images.shape}  (batch=4, channels=1, 28×28)")
    print(f"  Output shape: {output.shape}  (batch=4, 10 classes)")
    
    # Show predictions for fake images
    probs = torch.softmax(output, dim=1)
    _, predicted = torch.max(probs, 1)
    print(f"  Predicted digits: {predicted.tolist()}  (random — model is untrained)")
    
    print(f"\n--- How CNN layers work ---")
    print(f"  Conv2d:    slides small filters across image → detects edges, textures")
    print(f"  ReLU:      activation function → introduces non-linearity")
    print(f"  MaxPool2d: shrinks image by half → keeps strongest features")
    print(f"  Flatten:   converts 2D feature maps → 1D vector for classification")
    print(f"  Linear:    standard neural network layers → final classification")
    print(f"\n  Layer 1 learns: edges, simple textures")
    print(f"  Layer 2 learns: shapes, patterns (built from edges)")
    print(f"  FC layers:      combine patterns → 'this looks like a 7'")

except ImportError:
    print("PyTorch not installed. Run: pip install torch")
    print("Or open this notebook in Google Colab (pre-installed).")
````

## 3. Types of Programming Languages

### Compiled vs. Interpreted

| Feature | Compiled | Interpreted |
|---|---------|------------|
| **How** | Source → compiler → machine code → runs | Source → interpreter reads line by line |
| **Speed** | Faster (pre-compiled) | Slower (interpreted at runtime) |
| **Errors** | Caught at compile time | Found when you run it |
| **Examples** | C, C++, Go, Rust, Java* | **Python**, JavaScript, Ruby, R |

\* Java is hybrid: compiled to bytecode, then JVM interprets/JIT-compiles it.

### Strongly Typed vs. Weakly Typed

| Feature | Strongly Typed | Weakly Typed |
|---|----------------|-------------|
| **Meaning** | Types enforced; can't mix | Auto-conversion happens |
| **Example** | `"5" + 3` → ERROR | `"5" + 3` → `"53"` (JS) |
| **Languages** | **Python**, Java, Go | JavaScript, PHP, Perl |

### Static vs. Dynamic Typing

| Feature | Static | Dynamic |
|---|--------|--------|
| **Meaning** | Variable types declared upfront | Types determined at runtime |
| **Example** | `int x = 5;` (Java) | `x = 5` then `x = "hello"` |
| **Languages** | Java, C, Go, TypeScript | **Python**, JavaScript, Ruby |

|-------|--------------|----------|

| **Multi-paradigm** | Supports both OOP and functional | **Python**, JavaScript, Scala |

### Platform Independence

| Approach | How It Works | Languages |
|----------|-------------|----------|

### Why Is Python Slow But Still #1?

> Python is the **glue, not the engine**:  
> - PyTorch matrix math → runs in C++ / CUDA (fast)  
> - NumPy operations → runs in C (fast)  
> - Database queries → database engine (fast)  
> - API calls → network I/O (waiting anyway)  
> - Python just **connects these pieces** — speed of glue doesn't matter  

### Thread Safety & the GIL — When Does It Matter?

Python has a **Global Interpreter Lock (GIL)** — only one thread can execute Python bytecode at a time.

| Scenario | Impact | Solution |
|----------|--------|---------|
| **CPU-bound** (math, training) | GIL hurts — threads don't parallelize | Use `multiprocessing` (separate processes) or NumPy/PyTorch (bypass GIL in C) |
| **I/O-bound** (API calls, DB queries, file reads) | GIL doesn't matter — threads wait for I/O anyway | Use `threading` or `asyncio` (FastAPI uses this!) |
| **Web servers** (handling requests) | GIL doesn't matter — each request is I/O-heavy | FastAPI + `uvicorn` with async/await |

> **For this class:** The GIL is rarely a problem because:  
> 1. Heavy compute (PyTorch, NumPy) runs in C/CUDA — bypasses the GIL  
> 2. API calls (OpenAI, database) are I/O-bound — async handles this  
> 3. FastAPI is async-native — handles thousands of concurrent requests  
>  
> **When it IS a problem:** Pure Python CPU loops processing millions of items. Fix: use `multiprocessing` or vectorized NumPy.

````python
# Function with type hints (good practice — documents intent)
def calculate_f1(precision: float, recall: float) -> float:
    """Calculate F1 score from precision and recall."""
    if precision + recall == 0:
        return 0.0
    return 2 * (precision * recall) / (precision + recall)

# Type hints are NOT enforced — they're documentation
f1 = calculate_f1(0.85, 0.90)
print(f"F1 Score: {f1:.4f}")

# This is something you'll actually use in P1!
print(f"\nPerfect model:  F1 = {calculate_f1(1.0, 1.0):.4f}")
print(f"High precision: F1 = {calculate_f1(0.95, 0.60):.4f}")
print(f"Balanced:       F1 = {calculate_f1(0.80, 0.80):.4f}")
````

### 7.5 Error Handling

````python
# Error handling with try/except
def safe_divide(a, b):
    """Divide a by b, handling errors gracefully."""
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        print("Error: Cannot divide by zero!")
        return None
    except TypeError as e:
        print(f"Error: Wrong types — {e}")
        return None

print(safe_divide(10, 3))     # Works fine
print(safe_divide(10, 0))     # ZeroDivisionError handled
print(safe_divide("10", 3))  # TypeError handled
````

### 7.5b Logging — Why `print()` Is NOT for Production

> In production, `print()` is **invisible** — it goes to stdout and disappears.  
> **Logging** gives you: timestamps, severity levels, file output, filtering.  
> Every job posting says "production-grade code" — logging is part of that.

````python
import logging
import sys

# ---- FIX for Jupyter: basicConfig won't work if root logger already has handlers ----
# In Jupyter, a root handler is auto-added. We must remove it first.
root = logging.getLogger()
root.handlers.clear()  # remove Jupyter's default handler

# Now configure logging (do this ONCE at the top of your app)
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s [%(levelname)s] %(name)s: %(message)s",
    datefmt="%Y-%m-%d %H:%M:%S",
    stream=sys.stdout,     # force output to stdout (visible in notebooks)
    force=True             # Python 3.8+: override existing config
)

# Create a logger for THIS module (best practice: use __name__)
logger = logging.getLogger(__name__)

# ---- The 5 log levels (from least to most severe) ----
# Level number:  DEBUG=10, INFO=20, WARNING=30, ERROR=40, CRITICAL=50
logger.debug("This is a DEBUG message — detailed diagnostic info")
logger.info("This is an INFO message — normal operation")
logger.warning("This is a WARNING message — something unexpected")
logger.error("This is an ERROR message — something failed")
logger.critical("This is a CRITICAL message — app might crash")

# Note: DEBUG doesn't show above because we set level=INFO
# Only INFO (20) and above are shown

# ---- Enable/disable log levels at runtime ----
print("\n--- Changing log level at runtime ---")
logger.setLevel(logging.DEBUG)     # now show DEBUG messages too
logger.debug("NOW you can see DEBUG (level lowered to DEBUG)")

logger.setLevel(logging.WARNING)   # only WARNING and above
logger.info("This INFO is now HIDDEN (level raised to WARNING)")
logger.warning("This WARNING still shows")

logger.setLevel(logging.INFO)      # reset to INFO
logger.info("Back to INFO level")

# ---- Package-level logging (like Java's logback per-package config) ----
print("\n--- Package-level logging (like Java) ---")

# Each logger has a HIERARCHY based on dot-separated names
# "myapp.api" is a child of "myapp" — inherits its level
api_logger = logging.getLogger("myapp.api")         # child
db_logger = logging.getLogger("myapp.database")      # child
parent_logger = logging.getLogger("myapp")            # parent

# Set parent to WARNING — all children inherit it
parent_logger.setLevel(logging.WARNING)
api_logger.info("This is HIDDEN — parent is WARNING")
api_logger.warning("This shows — WARNING meets threshold")

# Override ONE child to DEBUG (others stay at WARNING)
api_logger.setLevel(logging.DEBUG)
api_logger.debug("API logger: DEBUG enabled (overridden)")
db_logger.info("DB logger: still HIDDEN (inherits parent WARNING)")
db_logger.warning("DB logger: WARNING shows")

print("\n--- Key: logger hierarchy = package hierarchy ---")
print('logging.getLogger("myapp")           → parent')
print('logging.getLogger("myapp.api")       → child (inherits parent level)')
print('logging.getLogger("myapp.database")  → child (can override)')
print("Same concept as Java/logback per-package log levels.")
````

## 8. OOP — Object-Oriented Programming in Python

> "Everything in Python is an object" — even integers, strings, and functions.  
> Unlike Java where `int` is a primitive and `Integer` is an object, **Python has no primitives.**  
> OOP organizes code so teams can work on the same project.

### Class vs Object — The Key Distinction

| Concept | What It Is | Analogy | Example |
|---------|-----------|---------|---------|
| **Class** | A **blueprint** / template | Architectural drawing of a house | `class PredictionService:` |
| **Object** | A **specific instance** built from the blueprint | An actual house you can live in | `service = PredictionService("model.pkl")` |

> **Java devs:** Same as Java. `class Dog {}` is the blueprint. `Dog myDog = new Dog();` is the object.  
> Python just drops the `new` keyword: `my_dog = Dog()`

### What is `__init__`?

> `__init__` is Python's **constructor** — the method that runs when you create an object.  
> It's Java's `public ClassName(args) { }` constructor.

```python
# Java:
# public class PredictionService {
#     private String modelPath;
#     public PredictionService(String modelPath) {   // constructor
#         this.modelPath = modelPath;
#     }
# }

# Python equivalent:
class PredictionService:
    def __init__(self, model_path: str):     # constructor
        self.model_path = model_path         # self = this (Java)
```

| Python | Java | What It Does |
|--------|------|-------------|
| `__init__(self, ...)` | `public ClassName(...)` | Constructor — runs on object creation |
| `self` | `this` | Reference to the current object instance |
| `self.x = x` | `this.x = x` | Store data on the object |
| `__str__(self)` | `toString()` | String representation |
| `__len__(self)` | `.size()` | Length/size of the object |
| `__eq__(self, other)` | `.equals()` | Equality check |

> The `__double_underscore__` methods are called **dunder methods** (double-under).  
> They're Python's way of implementing operator overloading and standard interfaces.

````python
# PROOF: Everything in Python is an object — even "primitives"

# Integers have methods (try this in Java with int!)
x = 42
print(f"x = {x}")
print(f"  type(x)        = {type(x)}")           # <class 'int'>
print(f"  id(x)          = {id(x)}")              # memory address
print(f"  x.bit_length() = {x.bit_length()}")     # int method!
print(f"  x.__add__(8)   = {x.__add__(8)}")       # 42 + 8 via method
print(f"  x.__class__    = {x.__class__}")         # it's a class instance

# Strings have methods
s = "hello"
print(f"\ns = '{s}'")
print(f"  type(s)      = {type(s)}")
print(f"  s.upper()    = {s.upper()}")
print(f"  s.__len__()  = {s.__len__()}")           # len(s) calls this

# Booleans are SUBCLASSES of int!
b = True
print(f"\nb = {b}")
print(f"  type(b)            = {type(b)}")
print(f"  isinstance(b, int) = {isinstance(b, int)}")  # True!
print(f"  True + True        = {True + True}")          # 2 (bools are ints)

# Even functions are objects
def greet(name):
    return f"Hello, {name}"

print(f"\ngreet is an object:")
print(f"  type(greet)      = {type(greet)}")
print(f"  id(greet)        = {id(greet)}")
print(f"  greet.__name__   = {greet.__name__}")
print(f"  greet.__class__  = {greet.__class__}")

# You can assign functions to variables, pass them around
say_hello = greet        # assign function to variable
print(f"  say_hello('AI') = {say_hello('AI')}")

print("\n--- Java: int is NOT an object. Integer IS an object. Two types. ---")
print("--- Python: int IS an object. Period. One type. Everything is. ---")
````

````python
# CLASS vs OBJECT — and why "load once" matters

# ---- WITHOUT OOP: function reloads every time ----
import time

def predict_v1(features, model_path):
    """Loads model every time — wasteful!"""
    print(f"  [LOADING] model from {model_path}...")  # expensive!
    time.sleep(0.01)  # simulate load time
    return {"prediction": "approved", "confidence": 0.87}

print("=== WITHOUT OOP: calling 5 times ===")
start = time.time()
for i in range(5):
    predict_v1({"age": 30}, "model.pkl")
elapsed_v1 = time.time() - start
print(f"  → Model loaded 5 times! Elapsed: {elapsed_v1:.3f}s\n")

# ---- WITH OOP: class loads once, predicts many ----
class PredictionService:
    """Blueprint (class). Load model once, predict many times."""
    
    def __init__(self, model_path: str):
        """Constructor — runs ONCE when you create the object."""
        self.model_path = model_path
        self.load_count = 0
        print(f"  [LOADING] model from {model_path}...")  # only once!
        time.sleep(0.01)  # simulate load time
        self.load_count += 1
    
    def predict(self, features: dict) -> dict:
        """Predict — model already loaded, just use it."""
        return {"prediction": "approved", "confidence": 0.87}

print("=== WITH OOP: calling 5 times ===")
start = time.time()
service = PredictionService("model.pkl")  # ← object (instance of the class)
for i in range(5):
    service.predict({"age": 30})
elapsed_v2 = time.time() - start
print(f"  → Model loaded {service.load_count} time. Predicted 5 times. Elapsed: {elapsed_v2:.3f}s")

print(f"\n--- Class vs Object ---")
print(f"  PredictionService = CLASS (the blueprint)")
print(f"  service           = OBJECT (an instance built from the blueprint)")
print(f"  type(service)     = {type(service)}")
print(f"\n  You can create MULTIPLE objects from ONE class:")
service_a = PredictionService("model_a.pkl")
service_b = PredictionService("model_b.pkl")
print(f"  service_a.model_path = {service_a.model_path}")
print(f"  service_b.model_path = {service_b.model_path}")
print(f"  Same class, different objects, different data.")
````

````python
# The 4 Pillars of OOP

# 1. ENCAPSULATION — bundle data + behavior, hide internals
class BankAccount:
    def __init__(self, owner: str, balance: float):
        self.owner = owner
        self.__balance = balance  # private (double underscore)
    
    def deposit(self, amount: float):
        if amount <= 0:
            raise ValueError("Amount must be positive")
        self.__balance += amount
    
    def get_balance(self) -> float:
        return self.__balance

account = BankAccount("Alex", 10000)
account.deposit(5000)
print(f"Balance: ${account.get_balance():,.2f}")
# print(account.__balance)  # This would FAIL — it's private!
````

````python
# 2. INHERITANCE — child class inherits from parent
# 3. POLYMORPHISM — same method name, different behavior
# 4. ABSTRACTION — define WHAT, not HOW

from abc import ABC, abstractmethod

# Abstract base class (Abstraction)
class MLModel(ABC):
    """Any ML model must implement predict(). Can optionally override describe()."""
    
    @abstractmethod
    def predict(self, features: dict) -> dict:
        """Subclasses MUST implement this."""
        pass
    
    @abstractmethod
    def get_model_name(self) -> str:
        """Subclasses MUST implement this."""
        pass
    
    def describe(self) -> str:
        """Default implementation — subclasses CAN override, but don't have to.
        Like a non-abstract method in a Java abstract class."""
        return f"Model: {self.get_model_name()} (no custom description)"
    
    def log_prediction(self, features: dict, result: dict):
        """Another default method — shared behavior across all models."""
        print(f"  [{self.get_model_name()}] {features} → {result['prediction']}")

# Concrete classes (Inheritance + Polymorphism)
class RandomForestModel(MLModel):
    def predict(self, features: dict) -> dict:
        return {"prediction": "approved", "model": self.get_model_name()}
    
    def get_model_name(self) -> str:
        return "RandomForest"
    
    def describe(self) -> str:
        """Overrides the default describe() — custom description."""
        return "Random Forest: ensemble of decision trees, great for tabular data"

class NeuralNetModel(MLModel):
    def predict(self, features: dict) -> dict:
        return {"prediction": "denied", "model": self.get_model_name()}
    
    def get_model_name(self) -> str:
        return "NeuralNetwork"
    # NOTE: does NOT override describe() — gets the default implementation

# Polymorphism in action — same method, different behavior
models = [RandomForestModel(), NeuralNetModel()]
print("=== Polymorphism: same method, different behavior ===")
for model in models:
    result = model.predict({"age": 30})
    model.log_prediction({"age": 30}, result)   # shared default method

print("\n=== Default vs Overridden describe() ===")
for model in models:
    print(f"  {model.describe()}")

# PROVE: you can't instantiate an abstract class directly
print("\n=== Can't instantiate abstract class ===")
try:
    broken = MLModel()
except TypeError as e:
    print(f"  TypeError: {e}")
    print("  → Must implement ALL @abstractmethod methods first")

print("\n--- Summary ---")
print("@abstractmethod     = subclass MUST implement (like Java 'abstract void')")
print("Regular method      = default implementation, subclass CAN override")
print("                      (like a regular method in Java abstract class)")
````

````python
# Python is BOTH Object-Oriented AND Functional — same problem, two styles

scores = [85, 92, 78, 95, 60, 88, 73, 91]

# ---- OOP Style: class with methods ----
class ScoreAnalyzer:
    def __init__(self, scores):
        self.scores = scores
    
    def get_passing(self, threshold=80):
        return [s for s in self.scores if s >= threshold]
    
    def get_average(self):
        return sum(self.scores) / len(self.scores)

analyzer = ScoreAnalyzer(scores)
print("=== OOP Style ===")
print(f"Passing: {analyzer.get_passing()}")
print(f"Average: {analyzer.get_average():.1f}")

# ---- Functional Style: lambda, map, filter ----
print("\n=== Functional Style ===")

# Lambda = anonymous one-line function
double = lambda x: x * 2
print(f"lambda in variable: double(5) = {double(5)}")

# BETTER: invoke lambda directly without storing in a variable
# Wrap in parentheses and call immediately
print(f"lambda invoked directly: {(lambda x: x * 2)(5)}")
print(f"lambda with string:     {(lambda s: s.upper())('hello')}")
print(f"lambda with two args:   {(lambda a, b: a + b)(3, 7)}")

# Where lambdas shine: inline in function arguments (no variable needed)
# filter() — keep items matching a condition
passing = list(filter(lambda s: s >= 80, scores))
print(f"\nfilter (>=80):    {passing}")

# map() — transform every item
curved = list(map(lambda s: min(s + 5, 100), scores))
print(f"map (curve +5):   {curved}")

# sorted() with key function
by_score = sorted(scores, key=lambda s: -s)  # descending
print(f"sorted (desc):    {by_score}")

# Real-world: sort list of dicts by a field
students = [{"name": "Alex", "score": 85}, {"name": "Sam", "score": 92}, {"name": "Jordan", "score": 78}]
by_name = sorted(students, key=lambda s: s["name"])
print(f"sorted dicts:     {[s['name'] for s in by_name]}")

# ---- Pythonic Style: comprehensions (preferred!) ----
print("\n=== Pythonic Style (Comprehensions) ===")
passing = [s for s in scores if s >= 80]           # filter
curved = [min(s + 5, 100) for s in scores]         # map
avg = sum(scores) / len(scores)                    # reduce
print(f"Passing:  {passing}")
print(f"Curved:   {curved}")
print(f"Average:  {avg:.1f}")

print("\n--- Key Takeaway ---")
print("Java devs: Python lets you write OOP when structure helps (services, models)")
print("           AND functional when it's cleaner (data transforms, one-liners)")
print("           List comprehensions are the 'Pythonic' way — prefer them over map/filter")
print("\n--- Lambda: when to use each style ---")
print("  lambda in variable:   double = lambda x: x * 2     # just use def instead")
print("  lambda inline:        sorted(data, key=lambda x: x['score'])  # ← THIS is the right use")
print("  lambda direct invoke: (lambda x: x * 2)(5)         # rare, but useful in one-offs")
````

## 9. Naming Conventions & Code Style

### Linux vs Windows: Case Sensitivity

| OS | Case Sensitive? | What This Means |
|---|---|---|
| **Linux** | **Yes** | `MyFile.py` and `myfile.py` are DIFFERENT files |
| **Windows** | No | `MyFile.py` and `myfile.py` are the same file |
| **macOS** | Depends (usually No) | Default filesystem is case-insensitive |

> **Rule: Always use lowercase.** Your code will run on Linux servers (Docker, cloud).  
> If you develop on Windows/Mac with mixed case, it will break on Linux.

### Python Naming Conventions (PEP 8)

| What | Convention | Example | Bad Example |
|------|-----------|---------|------------|
| Variables | `snake_case` | `model_path` | `modelPath`, `ModelPath` |
| Functions | `snake_case` | `calculate_f1()` | `calculateF1()`, `CalculateF1()` |
| Classes | `PascalCase` | `PredictionService` | `prediction_service`, `predictionService` |
| Constants | `UPPER_SNAKE` | `MAX_RETRIES = 3` | `maxRetries`, `max_retries` |
| Files/Modules | `snake_case` | `data_processor.py` | `DataProcessor.py`, `dataProcessor.py` |
| Packages/Dirs | `lowercase` | `src/`, `tests/` | `SRC/`, `Tests/` |
| Private | `_prefix` | `_internal_method()` | |
| "Really" private | `__prefix` | `self.__balance` | |

> Java uses `camelCase` for methods (`calculateF1`).  
> Python uses `snake_case` (`calculate_f1`).  
> **When in Python, follow Python conventions.**

### AI/ML/DL/DS Libraries — What They Do, Use Cases, and Limitations

#### Data Processing

| Library | What It Does | Use Case | Limitations |

> **scikit-learn vs statsmodels:** scikit-learn is for *prediction* (accuracy matters). statsmodels is for *inference* (understanding relationships — "does age significantly affect salary?"). Many data scientists use both.

## 15. Unit Testing with pytest

> How do you KNOW your code works? **You test it.**  
> Every project in this program (P1-P6) will have tests.  
> "I wrote tests for every system I built" — puts you ahead of 90% of bootcamp grads.

````python
# The function we want to test
def calculate_f1(precision: float, recall: float) -> float:
    """Calculate F1 score from precision and recall."""
    if precision + recall == 0:
        return 0.0
    return 2 * (precision * recall) / (precision + recall)


# In a real project, these tests go in tests/test_evaluate.py
# and you run: pytest tests/ -v

# For now, let's run them inline:
def test_perfect_score():
    assert calculate_f1(1.0, 1.0) == 1.0

def test_zero_inputs():
    assert calculate_f1(0.0, 0.0) == 0.0

def test_known_value():
    result = calculate_f1(0.8, 0.6)
    expected = 2 * (0.8 * 0.6) / (0.8 + 0.6)
    assert abs(result - expected) < 0.001

def test_symmetric():
    """F1(p, r) should equal F1(r, p)."""
    assert calculate_f1(0.9, 0.7) == calculate_f1(0.7, 0.9)

# Run all tests
tests = [test_perfect_score, test_zero_inputs, test_known_value, test_symmetric]
for test in tests:
    try:
        test()
        print(f"  PASSED: {test.__name__}")
    except AssertionError as e:
        print(f"  FAILED: {test.__name__} — {e}")

print("\nIn real projects, you run: pytest tests/ -v")
print("pytest discovers test files automatically and gives nice output.")
````

## 19. SQL Quick Overview

> SQL is **pre-work** (~8-10 hrs self-paced).  
> You'll use it when needed: P3 (metadata), P5 (agent queries DB), P6 (monitoring).  
> Here's a quick taste — practice on SQLBolt (sqlbolt.com).

````python
# Python + SQL using sqlite3 (built-in, no install needed)
import sqlite3

# 1. Connect (creates file if not exists)
conn = sqlite3.connect(":memory:")  # in-memory database for demo
cursor = conn.cursor()
print("Connected to database")

# 2. Create table
cursor.execute("""
    CREATE TABLE students (
        id INTEGER PRIMARY KEY,
        name TEXT,
        marks INTEGER
    )
""")
print("Table created")

# 3. Insert data
cursor.execute("INSERT INTO students VALUES (1, 'Sam', 85)")
cursor.execute("INSERT INTO students VALUES (2, 'Alex', 90)")
cursor.execute("INSERT INTO students VALUES (3, 'John', 78)")
conn.commit()
print("Data inserted")

# 4. Query data
print("\nAll students:")
cursor.execute("SELECT * FROM students")
for row in cursor.fetchall():
    print(f"  {row}")

# 5. Filter
print("\nStudents with marks > 80:")
cursor.execute("SELECT name, marks FROM students WHERE marks > 80")
for row in cursor.fetchall():
    print(f"  {row[0]}: {row[1]}")

# 6. Aggregate
cursor.execute("SELECT AVG(marks) FROM students")
avg = cursor.fetchone()[0]
print(f"\nAverage marks: {avg:.1f}")

# 7. Update
cursor.execute("UPDATE students SET marks = 95 WHERE name = 'Sam'")
conn.commit()

print("\nAfter update (Sam → 95):")
cursor.execute("SELECT * FROM students")
for row in cursor.fetchall():
    print(f"  {row}")

conn.close()
print("\nConnection closed")
````
