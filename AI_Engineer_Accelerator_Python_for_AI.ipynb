## 3. Types of Programming Languages

### Compiled vs. Interpreted

| Feature | Compiled | Interpreted |
|---|---------|------------|
| **How** | Source → compiler → machine code → runs | Source → interpreter reads line by line |
| **Speed** | Faster (pre-compiled) | Slower (interpreted at runtime) |
| **Errors** | Caught at compile time | Found when you run it |
| **Examples** | C, C++, Go, Rust, Java* | **Python**, JavaScript, Ruby, R |

\* Java is hybrid: compiled to bytecode, then JVM interprets/JIT-compiles it.

### Strongly Typed vs. Weakly Typed

| Feature | Strongly Typed | Weakly Typed |
|---|----------------|-------------|
| **Meaning** | Types enforced; can't mix | Auto-conversion happens |
| **Example** | `"5" + 3` → ERROR | `"5" + 3` → `"53"` (JS) |
| **Languages** | **Python**, Java, Go | JavaScript, PHP, Perl |

### Static vs. Dynamic Typing

| Feature | Static | Dynamic |
|---|--------|--------|
| **Meaning** | Variable types declared upfront | Types determined at runtime |
| **Example** | `int x = 5;` (Java) | `x = 5` then `x = "hello"` |
| **Languages** | Java, C, Go, TypeScript | **Python**, JavaScript, Ruby |

### Object-Oriented or Not

| Style | What It Means | Languages |
|-------|--------------|----------|
| **OOP** | Classes/objects with data + behavior | Java, C#, C++ |
| **Functional** | Pure functions, immutable data | Haskell, Clojure, Scala |
| **Multi-paradigm** | Supports both OOP and functional | **Python**, JavaScript, Scala |

### Platform Independence

| Approach | How It Works | Languages |
|----------|-------------|----------|
| JVM | Bytecode runs on any OS with JVM | Java, Scala, Kotlin |
| Interpreter everywhere | Python interpreter on Win/Mac/Linux | **Python**, Ruby |
| Browser | Runs in any browser | JavaScript |

### Why Is Python Slow But Still #1?

> Python is the **glue, not the engine**:  
> - PyTorch matrix math → runs in C++ / CUDA (fast)  
> - NumPy operations → runs in C (fast)  
> - Database queries → database engine (fast)  
> - API calls → network I/O (waiting anyway)  
> - Python just **connects these pieces** — speed of glue doesn't matter  

### Thread Safety & the GIL — When Does It Matter?

Python has a **Global Interpreter Lock (GIL)** — only one thread can execute Python bytecode at a time.

| Scenario | Impact | Solution |
|----------|--------|---------|
| **CPU-bound** (math, training) | GIL hurts — threads don't parallelize | Use `multiprocessing` (separate processes) or NumPy/PyTorch (bypass GIL in C) |
| **I/O-bound** (API calls, DB queries, file reads) | GIL doesn't matter — threads wait for I/O anyway | Use `threading` or `asyncio` (FastAPI uses this!) |
| **Web servers** (handling requests) | GIL doesn't matter — each request is I/O-heavy | FastAPI + `uvicorn` with async/await |

> **For this class:** The GIL is rarely a problem because:  
> 1. Heavy compute (PyTorch, NumPy) runs in C/CUDA — bypasses the GIL  
> 2. API calls (OpenAI, database) are I/O-bound — async handles this  
> 3. FastAPI is async-native — handles thousands of concurrent requests  
>  
> **When it IS a problem:** Pure Python CPU loops processing millions of items. Fix: use `multiprocessing` or vectorized NumPy.

````python
# Function with type hints (good practice — documents intent)
def calculate_f1(precision: float, recall: float) -> float:
    """Calculate F1 score from precision and recall."""
    if precision + recall == 0:
        return 0.0
    return 2 * (precision * recall) / (precision + recall)

# Type hints are NOT enforced — they're documentation
f1 = calculate_f1(0.85, 0.90)
print(f"F1 Score: {f1:.4f}")

# This is something you'll actually use in P1!
print(f"\nPerfect model:  F1 = {calculate_f1(1.0, 1.0):.4f}")
print(f"High precision: F1 = {calculate_f1(0.95, 0.60):.4f}")
print(f"Balanced:       F1 = {calculate_f1(0.80, 0.80):.4f}")
````

### 7.5 Error Handling

````python
# Error handling with try/except
def safe_divide(a, b):
    """Divide a by b, handling errors gracefully."""
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        print("Error: Cannot divide by zero!")
        return None
    except TypeError as e:
        print(f"Error: Wrong types — {e}")
        return None

print(safe_divide(10, 3))     # Works fine
print(safe_divide(10, 0))     # ZeroDivisionError handled
print(safe_divide("10", 3))  # TypeError handled
````

### 7.5b Logging — Why `print()` Is NOT for Production

> In production, `print()` is **invisible** — it goes to stdout and disappears.  
> **Logging** gives you: timestamps, severity levels, file output, filtering.  
> Every job posting says "production-grade code" — logging is part of that.

````python
import logging

# Configure logging (do this ONCE at the top of your app)
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s [%(levelname)s] %(name)s: %(message)s",
    datefmt="%Y-%m-%d %H:%M:%S"
)

# Create a logger for THIS module (best practice: use __name__)
logger = logging.getLogger(__name__)

# ---- The 5 log levels (from least to most severe) ----
logger.debug("This is a DEBUG message — detailed diagnostic info")
logger.info("This is an INFO message — normal operation")
logger.warning("This is a WARNING message — something unexpected")
logger.error("This is an ERROR message — something failed")
logger.critical("This is a CRITICAL message — app might crash")

# Note: DEBUG doesn't show above because we set level=INFO
# Only INFO and above are shown
````

## 8. OOP — Object-Oriented Programming in Python

> "Everything in Python is an object" — even integers, strings, and functions.  
> Unlike Java where `int` is a primitive and `Integer` is an object, **Python has no primitives.**  
> OOP organizes code so teams can work on the same project.

````python
# PROOF: Everything in Python is an object — even "primitives"

# Integers have methods (try this in Java with int!)
x = 42
print(f"x = {x}")
print(f"  type(x)        = {type(x)}")           # <class 'int'>
print(f"  id(x)          = {id(x)}")              # memory address
print(f"  x.bit_length() = {x.bit_length()}")     # int method!
print(f"  x.__add__(8)   = {x.__add__(8)}")       # 42 + 8 via method
print(f"  x.__class__    = {x.__class__}")         # it's a class instance

# Strings have methods
s = "hello"
print(f"\ns = '{s}'")
print(f"  type(s)      = {type(s)}")
print(f"  s.upper()    = {s.upper()}")
print(f"  s.__len__()  = {s.__len__()}")           # len(s) calls this

# Booleans are SUBCLASSES of int!
b = True
print(f"\nb = {b}")
print(f"  type(b)            = {type(b)}")
print(f"  isinstance(b, int) = {isinstance(b, int)}")  # True!
print(f"  True + True        = {True + True}")          # 2 (bools are ints)

# Even functions are objects
def greet(name):
    return f"Hello, {name}"

print(f"\ngreet is an object:")
print(f"  type(greet)      = {type(greet)}")
print(f"  id(greet)        = {id(greet)}")
print(f"  greet.__name__   = {greet.__name__}")
print(f"  greet.__class__  = {greet.__class__}")

# You can assign functions to variables, pass them around
say_hello = greet        # assign function to variable
print(f"  say_hello('AI') = {say_hello('AI')}")

print("\n--- Java: int is NOT an object. Integer IS an object. Two types. ---")
print("--- Python: int IS an object. Period. One type. Everything is. ---")
````

````python
# Step 1: Without OOP (how beginners write it)
def predict_v1(features, model_path):
    """Loads model every time — wasteful!"""
    print(f"Loading model from {model_path}...")  # expensive!
    # model = load(model_path)
    return {"prediction": "approved", "confidence": 0.87}

# Called 1000 times = loads model 1000 times!
result = predict_v1({"age": 30}, "model.pkl")
print(result)
````

````python
# Step 2: With OOP (how production code works)
class PredictionService:
    """Load model once, predict many times."""
    
    def __init__(self, model_path: str):
        """Constructor — runs once when you create the object."""
        self.model_path = model_path
        print(f"Loading model from {model_path}...")  # only once!
        # self.model = load(model_path)
    
    def predict(self, features: dict) -> dict:
        """Predict — model already loaded, just use it."""
        return {
            "prediction": "approved",
            "confidence": 0.87,
            "model": self.model_path
        }

# Create once
service = PredictionService("model.pkl")

# Call 1000 times — model loaded only once!
print(service.predict({"age": 30}))
print(service.predict({"age": 45}))
````

````python
# The 4 Pillars of OOP

# 1. ENCAPSULATION — bundle data + behavior, hide internals
class BankAccount:
    def __init__(self, owner: str, balance: float):
        self.owner = owner
        self.__balance = balance  # private (double underscore)
    
    def deposit(self, amount: float):
        if amount <= 0:
            raise ValueError("Amount must be positive")
        self.__balance += amount
    
    def get_balance(self) -> float:
        return self.__balance

account = BankAccount("Alex", 10000)
account.deposit(5000)
print(f"Balance: ${account.get_balance():,.2f}")
# print(account.__balance)  # This would FAIL — it's private!
````

````python
# 2. INHERITANCE — child class inherits from parent
# 3. POLYMORPHISM — same method name, different behavior
# 4. ABSTRACTION — define WHAT, not HOW

from abc import ABC, abstractmethod

# Abstract base class (Abstraction)
class MLModel(ABC):
    """Any ML model must implement predict()."""
    
    @abstractmethod
    def predict(self, features: dict) -> dict:
        pass
    
    @abstractmethod
    def get_model_name(self) -> str:
        pass

# Concrete classes (Inheritance + Polymorphism)
class RandomForestModel(MLModel):
    def predict(self, features: dict) -> dict:
        return {"prediction": "approved", "model": self.get_model_name()}
    
    def get_model_name(self) -> str:
        return "RandomForest"

class NeuralNetModel(MLModel):
    def predict(self, features: dict) -> dict:
        return {"prediction": "denied", "model": self.get_model_name()}
    
    def get_model_name(self) -> str:
        return "NeuralNetwork"

# Polymorphism in action — same method, different behavior
models = [RandomForestModel(), NeuralNetModel()]
for model in models:
    result = model.predict({"age": 30})
    print(f"{result['model']}: {result['prediction']}")
````

````python
# Python is BOTH Object-Oriented AND Functional — same problem, two styles

scores = [85, 92, 78, 95, 60, 88, 73, 91]

# ---- OOP Style: class with methods ----
class ScoreAnalyzer:
    def __init__(self, scores):
        self.scores = scores
    
    def get_passing(self, threshold=80):
        return [s for s in self.scores if s >= threshold]
    
    def get_average(self):
        return sum(self.scores) / len(self.scores)

analyzer = ScoreAnalyzer(scores)
print("=== OOP Style ===")
print(f"Passing: {analyzer.get_passing()}")
print(f"Average: {analyzer.get_average():.1f}")

# ---- Functional Style: lambda, map, filter ----
print("\n=== Functional Style ===")

# Lambda = anonymous one-line function
double = lambda x: x * 2
print(f"lambda double(5): {double(5)}")

# filter() — keep items matching a condition
passing = list(filter(lambda s: s >= 80, scores))
print(f"filter (>=80):    {passing}")

# map() — transform every item
curved = list(map(lambda s: min(s + 5, 100), scores))
print(f"map (curve +5):   {curved}")

# sorted() with key function
by_score = sorted(scores, key=lambda s: -s)  # descending
print(f"sorted (desc):    {by_score}")

# ---- Pythonic Style: comprehensions (preferred!) ----
print("\n=== Pythonic Style (Comprehensions) ===")
passing = [s for s in scores if s >= 80]           # filter
curved = [min(s + 5, 100) for s in scores]         # map
avg = sum(scores) / len(scores)                    # reduce
print(f"Passing:  {passing}")
print(f"Curved:   {curved}")
print(f"Average:  {avg:.1f}")

print("\n--- Key Takeaway ---")
print("Java devs: Python lets you write OOP when structure helps (services, models)")
print("           AND functional when it's cleaner (data transforms, one-liners)")
print("           List comprehensions are the 'Pythonic' way — prefer them over map/filter")
````

## 9. Naming Conventions & Code Style

### Linux vs Windows: Case Sensitivity

| OS | Case Sensitive? | What This Means |
|---|---|---|
| **Linux** | **Yes** | `MyFile.py` and `myfile.py` are DIFFERENT files |
| **Windows** | No | `MyFile.py` and `myfile.py` are the same file |
| **macOS** | Depends (usually No) | Default filesystem is case-insensitive |

> **Rule: Always use lowercase.** Your code will run on Linux servers (Docker, cloud).  
> If you develop on Windows/Mac with mixed case, it will break on Linux.

### Python Naming Conventions (PEP 8)

| What | Convention | Example | Bad Example |
|------|-----------|---------|------------|
| Variables | `snake_case` | `model_path` | `modelPath`, `ModelPath` |
| Functions | `snake_case` | `calculate_f1()` | `calculateF1()`, `CalculateF1()` |
| Classes | `PascalCase` | `PredictionService` | `prediction_service`, `predictionService` |
| Constants | `UPPER_SNAKE` | `MAX_RETRIES = 3` | `maxRetries`, `max_retries` |
| Files/Modules | `snake_case` | `data_processor.py` | `DataProcessor.py`, `dataProcessor.py` |
| Packages/Dirs | `lowercase` | `src/`, `tests/` | `SRC/`, `Tests/` |
| Private | `_prefix` | `_internal_method()` | |
| "Really" private | `__prefix` | `self.__balance` | |

> Java uses `camelCase` for methods (`calculateF1`).  
> Python uses `snake_case` (`calculate_f1`).  
> **When in Python, follow Python conventions.**

## 15. Unit Testing with pytest

> How do you KNOW your code works? **You test it.**  
> Every project in this program (P1-P6) will have tests.  
> "I wrote tests for every system I built" — puts you ahead of 90% of bootcamp grads.

````python
# The function we want to test
def calculate_f1(precision: float, recall: float) -> float:
    """Calculate F1 score from precision and recall."""
    if precision + recall == 0:
        return 0.0
    return 2 * (precision * recall) / (precision + recall)


# In a real project, these tests go in tests/test_evaluate.py
# and you run: pytest tests/ -v

# For now, let's run them inline:
def test_perfect_score():
    assert calculate_f1(1.0, 1.0) == 1.0

def test_zero_inputs():
    assert calculate_f1(0.0, 0.0) == 0.0

def test_known_value():
    result = calculate_f1(0.8, 0.6)
    expected = 2 * (0.8 * 0.6) / (0.8 + 0.6)
    assert abs(result - expected) < 0.001

def test_symmetric():
    """F1(p, r) should equal F1(r, p)."""
    assert calculate_f1(0.9, 0.7) == calculate_f1(0.7, 0.9)

# Run all tests
tests = [test_perfect_score, test_zero_inputs, test_known_value, test_symmetric]
for test in tests:
    try:
        test()
        print(f"  PASSED: {test.__name__}")
    except AssertionError as e:
        print(f"  FAILED: {test.__name__} — {e}")

print("\nIn real projects, you run: pytest tests/ -v")
print("pytest discovers test files automatically and gives nice output.")
````

## 19. SQL Quick Overview

> SQL is **pre-work** (~8-10 hrs self-paced).  
> You'll use it when needed: P3 (metadata), P5 (agent queries DB), P6 (monitoring).  
> Here's a quick taste — practice on SQLBolt (sqlbolt.com).

````python
# Python + SQL using sqlite3 (built-in, no install needed)
import sqlite3

# 1. Connect (creates file if not exists)
conn = sqlite3.connect(":memory:")  # in-memory database for demo
cursor = conn.cursor()
print("Connected to database")

# 2. Create table
cursor.execute("""
    CREATE TABLE students (
        id INTEGER PRIMARY KEY,
        name TEXT,
        marks INTEGER
    )
""")
print("Table created")

# 3. Insert data
cursor.execute("INSERT INTO students VALUES (1, 'Sam', 85)")
cursor.execute("INSERT INTO students VALUES (2, 'Alex', 90)")
cursor.execute("INSERT INTO students VALUES (3, 'John', 78)")
conn.commit()
print("Data inserted")

# 4. Query data
print("\nAll students:")
cursor.execute("SELECT * FROM students")
for row in cursor.fetchall():
    print(f"  {row}")

# 5. Filter
print("\nStudents with marks > 80:")
cursor.execute("SELECT name, marks FROM students WHERE marks > 80")
for row in cursor.fetchall():
    print(f"  {row[0]}: {row[1]}")

# 6. Aggregate
cursor.execute("SELECT AVG(marks) FROM students")
avg = cursor.fetchone()[0]
print(f"\nAverage marks: {avg:.1f}")

# 7. Update
cursor.execute("UPDATE students SET marks = 95 WHERE name = 'Sam'")
conn.commit()

print("\nAfter update (Sam → 95):")
cursor.execute("SELECT * FROM students")
for row in cursor.fetchall():
    print(f"  {row}")

conn.close()
print("\nConnection closed")
````
